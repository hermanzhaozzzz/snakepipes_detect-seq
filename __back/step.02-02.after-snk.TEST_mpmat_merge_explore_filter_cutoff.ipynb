```python
import os
import pandas as pd
# import seaborn as sns
```

```python
PATH = "../mpmat_filter/"
ls = [i for i in os.listdir(PATH) if '.mpmat' in i and 'test' not in i]
ls.sort()
ls[:4]
```

```python
# SiteMutNum = [1,2,3,4,5,6,7,8] # 师兄的M4 建议多跑几个range
# SiteCoverNum = [5,6,8,10,15,20] # 师兄的C10
# SiteMutRatio = [0.01,0.1] # 影响不大
# RegionPassNum = [1,2,3] # 师兄的R2
# RegionToleranceNum = [2] # 影响不大，师兄建议设置成2
ls_all = []
for i in range(len(ls)):
    # path
    path = ls[i]
    Mut_type = ls[i][:-6].split('.')[-1].replace('2','')
    Sample = ls[i].replace('293T-bat_','').split('_hg38')[0]
    # SMN
    Mut_num = int(ls[i].split('SMN-')[1].split('_')[0])
    # SCN
    Cover_num = int(ls[i].split('SCN-')[1].split('_')[0])
    # SMR
    Mut_ratio = float(ls[i].split('SMR-')[1].split('_')[0])
    # RPN
    Pass_num = int(ls[i].split('RPN-')[1].split('_')[0])
    # lines
    with open(os.path.join('../mpmat_filter',ls[i])) as f:
        print(ls[i])
        Count = len(list(f.readlines()))
    ls_all.append([Sample,Mut_type,Mut_num,Cover_num,Mut_ratio,Pass_num,Count])
ls_all[:4]
```

```python
df = pd.DataFrame(ls_all,columns=['Sample',"Mut_type","Mut_num","Cover_num","Mut_ratio","Pass_num","Count"])
df
```

```python
df_final = df[(df['Mut_type']=='CT') | (df['Mut_type']=='GA')]
df_final
```
