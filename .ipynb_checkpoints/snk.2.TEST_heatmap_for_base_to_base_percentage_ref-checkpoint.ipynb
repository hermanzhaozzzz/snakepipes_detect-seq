```python
import os
import math
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def parse_vcf2list(workdir):
    ls_all = []
    for f_stat in [ i for i in os.listdir(workdir) if '.stats' in i]:
        bam_ = f_stat.split('_')[0].replace('293T-RNASeq-','')
        stat_ = os.path.join(workdir,f_stat)
        f = open(stat_,'r')
        ls = [ i.strip().replace('>','\t').split('\t')[-3:] for i in f.readlines() if i[:2] == 'ST']
        if len(ls) != 12:
            print('error')
        f.close()
        ls_all.append([bam_,ls])
    ls_all.sort()
    return ls_all
```

```python
def parse_list2df(ls_all):
    ls_plot = []
    for index_ls,ls in enumerate(ls_all):
        name_bam = ls[0]
        df = pd.DataFrame(ls[1], columns=['From Base', 'To Base', 'Count'])
    
    
    
        df['log2(count+1)'] = df.Count.map(lambda x: math.log(int(x)+1,2))



        df.Count = df.Count.map(int)
        sum_ = df.Count.sum()
        df['perc'] = df.Count/sum_


        fb_s = ['A','C','G','T'][::-1]
        tb_s = ['A','C','G','T']
        df_count = pd.DataFrame(np.zeros(16).reshape(4,4),index=fb_s,columns=tb_s)
        df_log2count = pd.DataFrame(np.zeros(16).reshape(4,4),index=fb_s,columns=tb_s)
        df_percentage = pd.DataFrame(np.zeros(16).reshape(4,4),index=fb_s,columns=tb_s)

        for index in fb_s:
            for column in tb_s:
                if index == column:
                    df_count.loc[index,column] = np.NaN
                    df_log2count.loc[index,column] = np.NaN
                    df_percentage.loc[index,column] = np.NaN
                else:
                    df_count.loc[index,column] = df[(df['From Base']==index) & (df['To Base']==column)]['Count'].values
                    df_log2count.loc[index,column] = df[(df['From Base']==index) & (df['To Base']==column)]['log2(count+1)'].values
                    df_percentage.loc[index,column] = df[(df['From Base']==index) & (df['To Base']==column)]['perc'].values * 100
        df_count = pd.DataFrame(df_count.fillna(0),dtype='int64')
        col = index_ls%4
        row = int(index_ls/4)
        dt_plot = {
            "col": col,
            "row": row,
            "df_log": df_log2count,
            "df_count": df_count,
            "df_perc": df_percentage
        }
        ls_plot.append([name_bam,dt_plot])
    ls_plot.sort()
    return ls_plot
```

```python
def plot_log(ls_plot,coverage=100,ratio=0.01,number=4):
    total_row = ls_plot[-1][1]['row']+1
    fig,ax = plt.subplots(total_row,4,figsize=(4*4,4*total_row))
    for ls in ls_plot:
        dt = ls[1]
        name_bam = ls[0]
        df_log = dt['df_log']
        df_count = dt['df_count']
        row = dt['row']
        col = dt['col']
        sns.heatmap(
            df_log, 
            vmin = 5,
            vmax = 9,
            linewidths = 0.5,
            square = True,
            annot= df_count,
    #         annot= True,
            fmt = "d",
            ax = ax[row,col],
            cmap = 'Blues',
            cbar_kws={"orientation":"vertical","label": "log2(count+1)"}
        )
        ax[row,col].set_title(name_bam,fontsize=13,fontfamily = 'sans-serif',fontstyle='italic')
        ax[row,col].set_xlabel('To Base', fontsize=13,fontfamily = 'sans-serif',fontstyle='italic')
        ax[row,col].set_ylabel('From Base', fontsize=13,fontfamily = 'sans-serif',fontstyle='italic')
        
    for i in range(1,4):
        ax[3,i].spines['top'].set_visible(False)
        ax[3,i].spines['right'].set_visible(False)
        ax[3,i].spines['bottom'].set_visible(False)
        ax[3,i].spines['left'].set_visible(False)
        ax[3,i].set_xticks([])
        ax[3,i].set_yticks([])
        
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=0.5)
    fig.suptitle('Log2(count+1) Cutoff: coverage-{cov}, ratio-{ratio}, number-{number}'.format(cov=coverage,ratio=ratio,number=number), fontsize=16,fontfamily = 'sans-serif',fontstyle='italic')
    try:
        os.makedirs("../plot_heatmap")
    except:
        pass
    plt.savefig('../plot_heatmap/cutoff_{cov}_{ratio}_{number}_log.pdf'.format(cov=coverage,ratio=ratio,number=number))
#     plt.show()
def plot_perc(ls_plot,coverage=100,ratio=0.01,number=4):
    total_row = ls_plot[-1][1]['row']+1
    fig,ax = plt.subplots(total_row,4,figsize=(4*4,4*total_row))
    for ls in ls_plot:
        dt = ls[1]
        name_bam = ls[0]
        df_percentage = dt['df_perc']
        row = dt['row']
        col = dt['col']
        
        sns.heatmap(
            df_percentage, 
            vmin = 15,
            vmax = 25,
            linewidths = 0.5,
            square = True,
            annot= True,
            fmt = ".1f",
            ax = ax[row,col],
            cmap = 'Blues',
            cbar_kws={"orientation":"vertical","label": "Percentage of all"}
        )
        ax[row,col].set_title(name_bam,fontsize=13,fontfamily = 'sans-serif',fontstyle='italic')
        ax[row,col].set_xlabel('To Base', fontsize=13,fontfamily = 'sans-serif',fontstyle='italic')
        ax[row,col].set_ylabel('From Base', fontsize=13,fontfamily = 'sans-serif',fontstyle='italic')
    
    
    
    
    for i in range(1,4):
        ax[3,i].spines['top'].set_visible(False)
        ax[3,i].spines['right'].set_visible(False)
        ax[3,i].spines['bottom'].set_visible(False)
        ax[3,i].spines['left'].set_visible(False)
        ax[3,i].set_xticks([])
        ax[3,i].set_yticks([])
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=0.5)
    fig.suptitle('Percentage Cutoff: coverage-{cov}, ratio-{ratio}, number-{number}'.format(cov=coverage,ratio=ratio,number=number), fontsize=16,fontfamily = 'sans-serif',fontstyle='italic')
    try:
        os.makedirs("../plot_heatmap")
    except:
        pass
    plt.savefig('../plot_heatmap/cutoff_{cov}_{ratio}_{number}_perc.pdf'.format(cov=coverage,ratio=ratio,number=number))
#     plt.show()
```

```python
def main(coverage=100,ratio=0.2,number=4):
    workdir = "../vcf_bcftools_filt/{coverage}_{ratio}_{number}_filt/".format(coverage=coverage,ratio=ratio,number=number)
    ls = parse_vcf2list(workdir)
    ls_plot = parse_list2df(ls)
#     ls_plot = ls_plot[:4] + ls_plot[6:11]
    plot_log(ls_plot,coverage=coverage,ratio=ratio,number=number)
    plot_perc(ls_plot,coverage=coverage,ratio=ratio,number=number)
```
